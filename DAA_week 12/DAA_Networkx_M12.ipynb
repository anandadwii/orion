# Networkx
Networkx for network analytics

```python
import networkx as nx
import matplotlib.pyplot as plt
verticies = range(1,9)
edges = {(1,2),(2,3),(7,2),(4,5),(6,7),(2,5),(1,8),(1,5),(4,7),(7,8)}
G = nx.Graph()
G.add_nodes_from(verticies)
G.add_edges_from(edges)
nx.draw(G, with_labels=True, node_color='r', node_size=600)
```

```python
nx.degree_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

```python
nx.closeness_centrality(G)
```

```python
centrality = nx.eigenvector_centrality(G)
```

```python
sorted((v,'{:0.3f}'.format(c)) for v, c in centrality.items())
```

# facebook nodes analytics

```python
G_fb=nx.read_edgelist("facebook_combined.txt", create_using = nx.Graph(), nodetype=int)
```

```python
print(nx.info(G_fb))
```

```python
pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [2000.0*G_fb.degree(v) for v in G_fb]
node_size = [v* 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                node_color=node_color,
                node_size=node_size)
plt.axis('off')
```
